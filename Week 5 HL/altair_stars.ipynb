```python
data2 = pd.read_csv("stars(1).csv")
chart2 = alt.Chart(data2) # Chart prepares data into altair format

scatter2 =chart2.mark_circle().encode(x=alt.X("Temperature (K)",scale=alt.Scale(domain=[40000,0]),title="Surface temperature (K)"),
    y=alt.Y("Absolute magnitude (Mv)",scale=alt.Scale(domain=[25,-15])),
    color=alt.condition(brush,alt.Color("Star Type:N", scale=alt.Scale(scheme="category10")),
    alt.value("lightgray"))).add_selection(brush).properties(title=
    "A graph indicating the relationship between the surface temparture of a star and its absolute magnitude (brightness)"
    )

# mark_'object' used to specify type of graph, :N specifies an unordered discrete datatype
# encode used to specify how the data is presented (i.e. x,y axis, colour of point, shape of point)

scatter2

# This scatter graph shows that stars in the universe follow the Hertzsprung-Russell Diagram (see image attachment in GitHub folder)
```

```python
bars = chart2.mark_bar().encode(alt.X("Temperature (K)",bin=True,title="Surface temperature (K)"),y="count()",
    color=alt.Color("Star Type:N",scale=alt.Scale(scheme="category10")))

bars
```
